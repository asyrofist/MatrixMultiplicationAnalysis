### Charts

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 

 
# Set figure width to 12 and height to 9
fig_size[0] = 15
fig_size[1] = 15
plt.rcParams["figure.figsize"] = fig_size

# data
df=pd.read_csv('/mnt/final_results.csv')
df = df[df['n'] < 1025]
#df = df[df['Algorithm'] != 'divide-and-conquer']
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')

# plot
plt.plot('n', 'execution_time(s)', data=df[df['Algorithm']=='Strassen'], linestyle='-', marker='o', label='Strassen', color=palette(1), linewidth=4.0)
plt.plot('n', 'execution_time(s)',  data=df[df['Algorithm']=='ijk'], linestyle='-', marker='o', label='ijk', color=palette(2), linewidth=4.0)
plt.plot('n', 'execution_time(s)',  data=df[df['Algorithm']=='divide-and-conquer'], linestyle='-', marker='o', label='Divide and Conquer', color=palette(3), linewidth=4.0)

# Add legendplt.legend(loc=2, ncol=2)
plt.legend(loc=2, ncol=2)

# Add titles
plt.title("Tempo de Execução(s) x N", loc='center', fontsize=12, fontweight=2, color='navy')
plt.xlabel("N")
plt.ylabel("Tempo(s)")

plt.show()
```
